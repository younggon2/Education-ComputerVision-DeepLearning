## 클래스

```
# 연관성이 있는 함수들을 변수와 함께 지정
# 하나의 통합 기능을 클래스로 선언

class Arithmetic:
  def __init__(self, sum, subtract, multiply=True, divide=True):
    self.sum = sum
    self.subtract = subtract
    self.multiply = multiply
    self.divide = divide

  def get_sum(self, x, y):
    if self.sum:
      return x + y
    else:
      raise ValueError('Summation is not allowed.')

  def get_subtract(self, x, y):
    if self.subtract:
      return x - y
    else:
      raise ValueError('Subtraction is not allowed.')

  def get_multiply(self, x, y=5):
    if self.multiply:
      return x * y
    else:
      raise ValueError('Multiply is not allowed.')

  def get_divide(self, x, y):
    if self.divide:
      return x / y
    else:
      raise ValueError('Divide is not allowed.')

arithmetic_class = Arithmetic(sum=True, subtract=True, divide=False)
print(arithmetic_class.get_sum(1, 2))
print(arithmetic_class.get_subtract(1, 2))
print(arithmetic_class.get_multiply(2))
print(arithmetic_class.get_divide(6, 3))
```

```
# 부모 클래스를 상속받아 사용할 수 있음!

class ArithmeticChild(Arithmetic):
  def __init__(self, power):
    super(ArithmeticChild, self).__init__(sum=True, subtract=True, multiply=True, divide=True)
    self.power = power

  def get_power(self, x, y):
    if self.power:
      return x ** self.get_sum(x, y)
    else:
      raise ValueError('Power is not allowed.')

arithmeticchild_class = ArithmeticChild(True)
print(arithmeticchild_class.get_power(2, 3))
```

```
def plot_with_histogram(imglist, title=None):
  if title:
    assert len(imglist) == len(title)

  for i in range(len(imglist)):
    plt.figure(figsize=(15,7))
    plt.subplot(121)
    plt.imshow(imglist[i], 'gray')
    if title:
      plt.title(title[i])

    plt.subplot(122)
    plt.hist(imglist[i].flatten())
    if title:
      plt.title(title[i])
    plt.show()
```

```
# 이미지 경로 설정
image_path = 'test.jpg'  # 이미지 파일 경로를 입력하세요

# 이미지 읽기
img2 = cv2.imread(image_path)

# 필터 커널 크기 설정
kernel_size = 5  # 필터 커널 크기를 원하는 값으로 변경하세요

# 평균 필터 적용
mean_filtered = cv2.blur(img2, (kernel_size, kernel_size))

# 중간값 필터 적용
median_filtered = cv2.medianBlur(img2, kernel_size)

# 가우시안 필터 적용
gaussian_filtered = cv2.GaussianBlur(img2, (kernel_size, kernel_size), 0)

# 결과 출력
plt.figure(figsize=(16, 4))
plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(mean_filtered, cv2.COLOR_BGR2RGB))
plt.title('Mean Filtered')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(median_filtered, cv2.COLOR_BGR2RGB))
plt.title('Median Filtered')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(gaussian_filtered, cv2.COLOR_BGR2RGB))
plt.title('Gaussian Filtered')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```
# Salt & Pepper 노이즈 추가
noise_prob = 0.05  # 노이즈 비율 (0~1 사이의 값)
img_noise = img2.copy()
noise_mask = np.random.rand(*img2.shape[:2]) < noise_prob
salt_noise = noise_mask & (np.random.rand(*img2.shape[:2]) < 0.5)
pepper_noise = noise_mask & ~salt_noise
img_noise[salt_noise] = 0  # Salt 노이즈 (검정색)
img_noise[pepper_noise] = 255  # Pepper 노이즈 (흰색)


# 중간값 필터 적용
median_filtered = cv2.medianBlur(img_noise, 5)

# 결과 출력
plt.figure(figsize=(16, 4))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(img_noise, cv2.COLOR_BGR2RGB))
plt.title('Salt & Pepper Noise Added')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(median_filtered, cv2.COLOR_BGR2RGB))
plt.title('Median Filtered')
plt.axis('off')

plt.tight_layout()
plt.show()
```
