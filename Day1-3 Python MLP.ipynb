> linear gate와는 다르게 하나의 직선으로 0과 1을 구분지을 수 없다. 이 문제에 대한 해결방법을 생각 해 보자.


````python
import numpy as np

def AND_Perceptron(x1,x2):
    w = np.array([0.5,0.5])
    b = -0.7
    theta = 0

    x = np.array([x1,x2])
    y = np.sum(w*x)+b

    if y > theta:
        return 1
    elif y <= theta:
        return 0

def OR_Perceptron(x1,x2):
    w = np.array([0.5,0.5])
    b = 0
    theta = 0

    x = np.array([x1,x2])
    y = np.sum(w*x)+b

    if y > theta:
        return 1
    elif y <= theta:
        return 0

def NAND_Perceptron(x1,x2):
    w = np.array([0.5,0.5])
    b = 0
    theta = 0.5

    x = np.array([x1,x2])
    y = np.sum(w*x)+b

    if y <= theta:
        return 1
    elif y > theta:
        return 0

def XOR_Perceptron(x1,x2):
    l1_0 = OR_Perceptron(x1,x2)
    l1_1 = NAND_Perceptron(x1,x2)
    l2   = AND_Perceptron(l1_0,l1_1)
    return l2


# inputData = np.array([[0,0],[0,1],[1,0],[1,1]])
# Perceptrons = [AND_Perceptron,OR_Perceptron,NAND_Perceptron,XOR_Perceptron]
# for box in Perceptrons:
#     print(f"-{box}-")
#     for i in inputData:
#         print(str(i) + " -----> " + str(box(i[0],i[1])))
#     print("-"*16)
````

**※ Python을 통한 구현** </br>
```python
class Affine:

    def __init__(self, W, B):
        self.W = W
        self.B = B
        self.x = None
        self.dW = None
        self.db = None
        
    def forward(self, x):
        self.x = x
        out = np.dot(x, self.W) + self.B
        return out
    
    def backward(self, dout):
        dx = np.dot(dout, self.W.T)
        self.dW = np.dot(self.x.T, dout)
        self.db = np.sum(dout, axis=0)
        return dx
```

### Activation function
- 활성화 함수(Activation Function)는 신경학적으로 볼때 뉴런발사(Firing of a Neuron)의 과정</br>

- 최종출력 신호를 다음 뉴런으로 보내줄지 말지 결정하는 역할을 하게 된다.</br>

- 뉴런이 다음 뉴런으로 신호를 보낼 때 입력신호가 일정 기준 이상이면 보내고 기준에 달하지 못하면 보내지 않을 수도 있다. 그 신호를 결정 해주는 것이 활성화 함수(Activation Function)
- 많은 종류의 활성화 함수가 있고, Activation function의 결정이 결과에 큰 영향을 준다.

</br></br>
 **※ Linear function**</br></br>
 - 선형함수는 말 그대로 직선적인 함수(y=x)이다.

></br>

</br>

</br>
>3) 계산이 복잡하다 (exp계산에 연산량이 비싸다)</br></br></br></br>

def sigmoid(self,x):

def relu(self,x):
    self.bool = (x<=0)
    y = x.copy()

 def softmax(x):
   exp_x = np.exp(x)

### Loss function</br>
손실함수란? </br>
- 신경망의 성능을 나타내기 위한 함수로,  훈련데이터를 잘 처리하지 못했는가에 대한 척도가 된다.</br>
- 모델의 높은 정확도를 위해서는 backpropagation을 통해 손실값을 최대한 작게 하는 최적의 매개변수값을 찾아야 한다.</br>

1. Mean Squared Error(MSE)
가장 일반적으로 쓸 수 있는 손실함수로는 평균제곱오차가 있다.</br>
>예측값과 실제값의 차이가 커질수록 제곱 연산으로 인해 값이 뚜렷해지고 오차가 양수이든 음수이든 누적값을 증가시킨다.</br>
>반대로 1미만 값은 더 작아지고, 그 이상의 값은 더 커져 값의 왜곡이 있을 수 있다.</br>

### Training</br>

 - 퍼셉트론은 네트워크가 예측할 때 Loss function을 통해 만드는 오차를 반영하도록 하여 오차가 감소하는 방향으로 연결을 강화시킨다.

 - 즉, 모든 출력 뉴런에 대해 올바른 예측을 만들 수 있도록 아래 수식과 같이 입력에 연결된 가중치를 강화시킨다.

> $w_{ij}^{(next step)} = w_{ij} + lr(y_j - \hat{y}_j)x_i$
>
> $w_{ij}$는 i번째 입력뉴련과j번째 뉴런의 입력값
>
> $x_i$는 현재 훈련 샘플의 i번째 뉴런의 입력값
>
> $\hat{y}_i$는 현재 훈련 샘플의 j번째 출력 뉴런의 출력값
>
> $y_j$는 현재 훈련 샘플의 j번째 출력 뉴런의 타깃값
>
> $lr$ = 학습률

### Mnist 데이터셋을 활용한 분류기 만들기 (Classifier)

````python
import matplotlib.pyplot as plt
from tensorflow.keras.datasets.mnist import load_data
import numpy as np

(train_images, train_labels), (test_images,test_labels) = load_data()
base = train_images[0]
for i in range(1,20):
    img1 = train_images[i]
    base = np.hstack((base,img1))
plt.imshow(base)
````

````python
#Affine class 정의
class Affine:

    def __init__(self, W, B):
        self.W = W
        self.B = B
        self.x = None
        self.dW = None
        self.db = None

    def forward(self, x_):
        self.x = x_
        out = self.x @ self.W+ self.B
        return out

    def backward(self, dout_):
        dx = dout_ @ self.W.T
        self.dW = self.x.T@ dout_
        self.db = np.sum(dout_, axis=0)
        return dx

#Categorical cross entropy error class 정의
class Ccee():
  def __init__(self):
    self.params = []

  def forward(self,pred_,gt_):
    bsize = len(pred_)
    j = np.mean(-np.log(pred_[np.arange(bsize), gt_]))
    return j

  def backward(self,pred_,gt_):
    bsize = len(pred_)
    dJ = -1/bsize*(gt_ - pred_)
    return dJ

#Sigmoid class 정의
class Sigmoid():
    def __init__(self):
        self.params = []

    def forward(self, z_):
        out = 1 / (1 + np.exp(-z_))
        return out

    def backward(self, dout_,out_):
        dX = dout_ * out_ * (1 - out_)
        return dX

#Softmax function 정의
def softmax(l_):
  pred = np.exp(l_)/np.sum(np.exp(l_), axis=1, keepdims=True)
  return pred

#n_correct counting function 정의
def count(pred_,gt_,b_size):
  global n_correct_ ,n_data_
  Pred_label = np.argmax(pred_, axis=1)
  n_correct_ += np.sum(Pred_label == gt_)
  n_data_ += b_size
  return Pred_label,n_correct_,n_data_

````

````python
import numpy as np
from numpy.random import normal
from numpy import zeros
from termcolor import colored
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
plt.style.use('seaborn')

# set test env.
N_DATA=1000
N_DATA_TRAIN=500

# N_DATA = train_images.shape[0]
N_FEATURE = train_images.shape[1]*train_images.shape[2]
B_SIZE = 64# batch size
N_ITER = N_DATA // B_SIZE
N_ITER_t = N_DATA_TRAIN // B_SIZE
EPOCHS = 100
LR = 0.3
UNITS = [64, 32, 10]
FLAG_PRINT_SHOW = True

# initialize w, b
W1 = normal(0, 1, (N_FEATURE, UNITS[0]))
B1 = zeros((UNITS[0]))

W2 = normal(0, 1, (UNITS[0], UNITS[1]))
B2 = zeros((UNITS[1]))

W3 = normal(0, 1, (UNITS[1], UNITS[2]))
B3 = zeros((UNITS[2]))

layer1=Affine(W1,B1)
layer2=Affine(W2,B2)
layer3=Affine(W3,B3)
ccee=Ccee()
sigmoid=Sigmoid()
````

````python
#### training ####

train_losses, train_accs = list(), list()
test_losses, test_accs = list(), list()
for epoch in range(EPOCHS):
    n_correct_, n_data_ = 0, 0
    for b_idx in range(N_ITER):
                # get mini-batch
                images = train_images[b_idx*B_SIZE : (b_idx + 1)*B_SIZE, ...] # batch size에 맞춰 mini-batch 가져오기
                X = images.reshape(B_SIZE, -1) / 255.               # X :images
                Y = train_labels[b_idx*B_SIZE : (b_idx + 1)*B_SIZE] # Y :labels

                ### forward propagation ###
                # layer1
                Z1 = layer1.forward(X)
                A1 = sigmoid.forward(Z1)
                # layer2
                Z2 = layer2.forward(A1)
                A2 = sigmoid.forward(Z2)
                # layer3
                Z3 = layer3.forward(A2)
                # 최종 Ouput
                Pred = softmax(Z3)
                # loss 계산
                J = ccee.forward(Pred,Y)
                #loss 기록
                train_losses.append(J)
                # calculate accuracy #count correct number and n_data
                Pred_label,n_correct_,n_data_ = count(Pred,Y,B_SIZE)

                ### backpropagation ###
                # gt -> one-hot encoding
                labels = Y.copy()
                Y = np.zeros_like(Pred)
                Y[np.arange(B_SIZE), labels] = 1
                # loss bp
                dJ = ccee.backward(Pred,Y)
                # layer3 bp
                dA2 = layer3.backward(dJ)
                # layer2 bp
                dZ2 = sigmoid.backward(dA2,A2)
                dA1 = layer2.backward(dZ2)
                # layer1 bp
                dZ1 = sigmoid.backward(dA1,A1)
                dd  = layer1.backward(dZ1)
                # parameter update
                layer3.W, layer3.B = layer3.W-LR*layer3.dW, layer3.B-LR*layer3.db
                layer2.W, layer2.B = layer2.W-LR*layer2.dW, layer2.B-LR*layer2.db
                layer1.W, layer1.B = layer1.W-LR*layer1.dW, layer1.B-LR*layer1.db

    ## flag ==true : first & last epoch's weight visualization
    if FLAG_PRINT_SHOW == True:
        if (epoch == 0 or epoch == EPOCHS-1) :
            plt.figure(figsize=(32,8))
            for i in range(0,5):
                plt.subplot(1,5,i+1)
                tmp=layer3.W[i].reshape(2,5)
                tmp1= normalize(tmp, axis=1, norm='l2')
                plt.imshow(tmp1)
                plt.axis('off')
                plt.title(f'Unit{i+1}')
    else:
            plt.figure(figsize=(32,8))
            for i in range(0,5):
                plt.subplot(1,5,i+1)
                tmp=layer3.W[i].reshape(2,5)
                tmp1= normalize(tmp, axis=1, norm='l2')
                plt.imshow(tmp1)
                plt.axis('off')
                plt.title(f'Unit{i+1}')

#### Test ####
    n_correct_t, n_data_t = 0, 0
    for b_idx in range(N_ITER_t):
        # get mini-batch
            images = test_images[b_idx*B_SIZE : (b_idx + 1)*B_SIZE, ...]
            X = images.reshape(B_SIZE, -1) / 255.
            Y = test_labels[b_idx*B_SIZE : (b_idx + 1)*B_SIZE]

            ### forward propagation
            # layer1
            Z1 = layer1.forward(X)
            A1 =sigmoid.forward(Z1)
            # layer2
            Z2=layer2.forward(A1)
            A2 = sigmoid.forward(Z2)
            # layer3
            Z3=layer3.forward(A2)
            Pred = softmax(Z3)
            # loss
            J = ccee.forward(Pred,Y)
            #loss 기록
            test_losses.append(J)
            # calculate accuracy
            Pred_label = np.argmax(Pred, axis=1)
            n_correct_t += np.sum(Pred_label == Y)
            n_data_t += B_SIZE

    plt.show()
    t_loss_ = train_losses[-1:]
    # print('Sum of loss',np.sum(layer2.W))
    if epoch == 0:
      print(f'Epoch{epoch+1}:','train_acc:',n_correct_/N_DATA, f'train_loss: {t_loss_[0]:.4f}','test_acc:',n_correct_t/N_DATA_TRAIN)
    if epoch == ((EPOCHS)-1):
      print(f'Epoch{epoch+1}:','train_acc:',n_correct_/N_DATA, f'train_loss: {t_loss_[0]:.4f}','test_acc:',n_correct_t/N_DATA_TRAIN)

    train_accs.append(n_correct_/N_DATA)
    test_accs.append(n_correct_t/N_DATA_TRAIN)
````

````python
fig, axes = plt.subplots(2, 1, figsize=(15, 10))
axes[0].plot(train_losses,'skyblue')
axes[1].plot(train_accs,'b',label='train acc')
axes[1].plot(test_accs,'r',label='test acc')

axes[0].set_title("Train Loss", color='darkblue', fontsize=20)
axes[0].set_xlabel("Iter Idx", fontsize=20)
axes[0].set_ylabel("CCEE", fontsize=20)

axes[1].set_title("Train Accuracy", color='darkblue', fontsize=20)
axes[1].set_xlabel("Epoch", fontsize=20)
axes[1].set_ylabel("Accuarcy", fontsize=20)
axes[1].set_yticks(np.linspace(0.4, 1.0, 7))

axes[0].tick_params(labelsize=20)
axes[1].tick_params(labelsize=20)
plt.legend()
fig.tight_layout()
````
